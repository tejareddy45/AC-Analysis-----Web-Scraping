```python
Brand=[]
Cooling_Capacity=[]
Star_Rating=[]
Annual_power_usage=[]
Room_Size=[]
Warranty=[]
Price=[]
Ratings=[]
no_of_rating=[]
Reviews=[]
for i in range(1,30):
    url = f"https://www.flipkart.com/air-conditioners/pr?sid=j9e%2Cabm%2Cc54&p%5B%5D=facets.fulfilled_by%255B%255D%3DFlipkart%2BAssured&p%5B%5D=facets.technology%255B%255D%3DInverter&p%5B%5D=facets.serviceability%5B%5D%3Dtrue&otracker=categorytree&otracker=nmenu_sub_TVs+%26+Appliances_0_Inverter+AC&page={i}"
    
    page= requests.get(url, headers=request_header)
    
    soup = BeautifulSoup(page.text)
    
    #for i in soup.find_all("div",class_="tUxRFH"):
        #AC_Contains.append(i.text)
    for i in soup.find_all("div",class_="KzDlHZ"):
        text1 = i.text
        
        regex = re.findall(r"^([a-zA-Z]+)\s",text1)
        if regex:
            Brand.append(regex[0])
        else:
            Brand.append(np.nan)
            
        regex = re.findall(r"(\d+\.?\d*\s*Ton)", text1, re.IGNORECASE)
        if regex:
            Cooling_Capacity.append(regex[0])
        else:
            Cooling_Capacity.append(np.nan)
            
        regex = re.findall(r"\d{1,2}\s?Star",text1)
        if regex:
            Star_Rating.append(regex[0])
        else:
            Star_Rating.append(np.nan)
            
    for i in soup.find_all("div",class_ = "_6NESgJ"):
        text2= i.text
        regex=re.findall(r"\d+\.\d+",text2)
        if regex:
            Annual_power_usage.append(regex[0])
        else:
            Annual_power_usage.append(np.nan)
        regex=re.findall(r"\:\s(\d+\s\-?\s?\d?\d?\d?)",text2)
        if regex:
            Room_Size.append(regex[0])
        else:
            Room_Size.append(np.nan)
        regex=re.findall(r"t(\d)\sY",text2)
        if regex:
            Warranty.append(regex[0])
        else:
            Warranty.append(np.nan)
            
            
            
            
    for i in soup.find_all("div",class_ = "tUxRFH"):
        text3 = i.find('div', class_="Nx9bqj _4b5DiR")
        if text3:
            regex = re.findall(r"\₹(\d.+)",text3.text)
            if regex:
                Price.append(regex[0])
            else:
                Price.append(np.nan)
        else:
            Price.append(np.nan)
    
    for i in soup.find_all("div",class_ ="tUxRFH"):
        text4 = i.find("span",class_="Y1HWO0")
        #regex = re.findall(r"\d+\.d+",text)
        if text4:
            Ratings.append(text4.text)
        else:
            Ratings.append(np.nan)
        
    for i in soup.find_all("div",class_ ="tUxRFH"):
        text5 = i.find("span",class_="Wphh3N")
        if text5:
            regex = re.findall('(\d+?\,?\d+)\sRa', text5.text)
            if len(regex) > 0:
                no_of_rating.append(regex[0])
            else:
                no_of_rating.append(np.nan)
        else:
            no_of_rating.append(np.nan)
        if text5:
            regex = re.findall('\s(\d+\,?\d+?)\sRe', text5.text)
            if len(regex) > 0:
                Reviews.append(regex[0])
            else:
                Reviews.append(np.nan)
        else:
            Reviews.append(np.nan)

         
        
```
