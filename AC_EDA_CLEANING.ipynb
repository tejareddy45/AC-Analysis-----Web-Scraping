### Calculating the average room size

```python
def calculate_average_room_size(value):
    # Check if the room size is a range (contains ' - ')
    if isinstance(value, str) and ' - ' in value:
        start, end = value.split(' - ')
        return (float(start) + float(end)) / 2  # Average of the range
    else:
        return float(value)  # Single value

# Applying the function to the 'Room_Size' column to get average room size for each row
datac['Average_Room_Size'] = datac['Room_Size'].apply(calculate_average_room_size)

# Display the updated DataFrame with the new 'Average_Room_Size' column
datac = datac[['Brand', 'Cooling_Capacity', 'Star_Rating', 'Annual_power_usage', 'Room_Size', 'Average_Room_Size', 'Warranty', 'Price', 'Ratings', 'no_of_rating', 'Reviews']]
```
